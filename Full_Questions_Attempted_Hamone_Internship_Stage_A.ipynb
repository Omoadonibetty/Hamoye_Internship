```python
# Importing the necessary python libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import dataset for quiz
url = 'https://github.com/HamoyeHQ/HDSC-Introduction-to-Python-for-machine-learning/files/7768140/FoodBalanceSheets_E_Africa_NOFLAG.csv'
fao_data = pd.read_csv(url, encoding= 'latin-1')
fao_data
```

```python
#checking for duplicated value
fao_data.duplicated().sum()
```

```python
#Perform a groupby operation on ‘Element’.  What year has the highest sum of Stock Variation?
fao_data.groupby(['Element']).sum().loc['Stock Variation']['Y2014':].sort_values(ascending=False)
```

```python
#Unique countries
fao_data['Area'].nunique()
```

```python
#checking for the sum of all items
fao_data.groupby(['Item']).sum()
```

```python
# Number of columns with missing values by percentage

total_no = fao_data.isnull().sum().sort_values(ascending=False)

percent = fao_data.isnull().sum()/fao_data.isnull().count()*100
percent1 = (round(percent1, 2)).sort_values(ascending=False)

missing_values = pd.concat([total_no, percent1], axis=1, keys=['Total_Number', '%'])
missing_values.head()
```

```python
#year with the least correlation value
correlations = fao_data[['Element Code', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].corr().iloc[0, 1:]


least_correlated_year = correlations.idxmin()
correlation_value = correlations.min()

print(correlation_value)
print(least_correlated_year)
```

```python
#question 15
#Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the highest sum in 2017?
grouped = fao_data.groupby('Area')['Y2017'].sum()
grouped
area_highest_sum = grouped.idxmax()
highest_sum = grouped.max()
print(highest_sum)
print(area_highest_sum)
```

```python
Y2017_data = fao_data['Y2017']
mean_2017 = Y2017_data.mean()
std_2017 = Y2017_data.std()

print(f"Mean for the Y2017: {mean_2017:.2f}")
print(f"Standard deviation for the Y2017: {std_2017:.2f}")
```
